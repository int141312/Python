```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import pandas as pd
import requests

import warnings
warnings.filterwarnings('ignore')

from wordcloud import WordCloud
import matplotlib.pyplot as plt
# %matplotlib inline
plt.rcParams['font.family'] = 'Malgun Gothic'

#크롬 드라이버 인식
driver = webdriver.Chrome()

# 사이트 접속
driver.get('https://www.youtube.com/watch?v=kekXCV3O-30')

# 페이지 밑으로 내리기. 단, 이 방법은 시간을 요구하며 정확한 댓글 갯수 만큼을 이동하지 못한다. (최적화x)
for i in range(0,500):
    driver.find_element_by_css_selector('body').send_keys(Keys.PAGE_DOWN)
sleep(1)

# 리스트 생성
comment_list = []
like_list = []

# 댓글 내용 추출
comment = driver.find_elements(by=By.CSS_SELECTOR, value = '#content-text')
like = driver.find_elements(by=By.CSS_SELECTOR, value = '#vote-count-middle')

```

```python
# 긁어온 댓글들을 하나의 텍스트 형식으로 바꾸기
for i in comment:
    comment_list.append(i.text)

txt = ' '.join(comment_list)
txt
```

```python
# 워드클라우드 실행
wc=WordCloud(font_path='./NanumGothic.ttf',
             background_color='black',
             width=1000,height=1000,
             max_words=100,
             max_font_size=200)
wordFre=wc.generate(txt)


plt.figure(figsize=(12,12))
plt.imshow(wordFre,interpolation='bilinear')
plt.axis('off')
plt.show()
```
