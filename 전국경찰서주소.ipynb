```python
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from selenium.webdriver.common.action_chains import ActionChains
import re
import pandas as pd


import warnings
warnings.filterwarnings('ignore')
```

```python
#크롬 드라이버 인식
driver = webdriver.Chrome()

# 검색어
query = ['서울 경찰서', '부산 경찰서', '대구 경찰서', '인천 경찰서', '광주 경찰서', '대전 경찰서', '울산 경찰서', '세종 경찰서',
         '경기도 경찰서', '강원도 경찰서', '충북 경찰서', '충남 경찰서', '전북 경찰서', '전남 경찰서', '경북 경찰서', '경남 경찰서', '제주 경찰서']

# 전체를 담을 빈 데이터프레임 생성
df_total = pd.DataFrame({ '경찰서 이름': [], '종류': [], '주소' : [] })

for i in range(0, len(query)):
    # 네이버 지도 접속
    driver.get(f'https://map.naver.com/v5/search/{query[i]}/place/1817652623?c=14120264.2927673,4518382.0005602,10,0,0,0,dh')
    sleep(2)

    # 원하는 특정 엘리먼트 추출: 이름, 가격
    driver.switch_to.frame("searchIframe")

    count_num1 = 0
    count_num2 = 0
    cop_name_list = []
    address_list = []
    kind_list = []

    # 한 페이지만 진행. 시도별로 경찰서가 50개가 넘지 않기에 1페이지에 다 출력된다.

    # 첫 번째 경찰서 선택
    driver.find_element(by=By.XPATH, value= f'//*[@id="_pcmap_list_scroll_container"]/ul/li[1]/div[1]/a').click()
    sleep(1)

    # 마지막까지 스크롤 내리기
    for i in range(0,50):
        driver.find_element_by_css_selector('body').send_keys(Keys.PAGE_DOWN)
    sleep(2)

    # 이름, 종류, 주소 추출
    cop_name = driver.find_elements(by=By.CSS_SELECTOR, value = 'span.place_bluelink._3Apve')
    kind = driver.find_elements(by=By.CSS_SELECTOR, value = 'span._3B6hV')
    address = driver.find_elements(by=By.CSS_SELECTOR, value = 'span._2Po-x')
    
    for name in cop_name:
        count_num1 = count_num1+1
        print(f'{count_num1}. {name.text}')
        cop_name_list.append(name.text)
        
    for i in address:
        count_num2 = count_num2+1
        print(f'{count_num2}. {i.text}')
        address_list.append(i.text)    
    
    for i in kind:
        kind_list.append(i.text)

    # 데이터프레임으로 만들기       
    df_cop = pd.DataFrame({ '경찰서 이름':cop_name_list, '종류':kind_list, '주소' : address_list,})

    # 만들어진 데이터프레임 합치기
    df_total = pd.concat([df_total, df_cop])

    
df_total.reset_index(drop=True, inplace=True)    
df_total
```
